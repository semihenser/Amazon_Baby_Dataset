##5

4 ve 5 yıldızları seçersem onlar 1 diğer geri kalanlkar 0 alacak yani 1 ve 2 puanlar.

```python
products['sentiment'] = products['rating'] >= 4
```

##8


```python
sentiment_model = turicreate.logistic_classifier.create(train_data,target='sentiment', features=['kelime_sayaci'], validation_set=test_data)
```

```python
#sentiment_model = graphlab.logistic_classifier.create(train_data,target='sentiment', features=['kelime_sayaci'], validation_set=test_data)
```

##9

Sentiment_modeli eğitelim.

```python
evaluate_roc=sentiment_model.evaluate(test_data,metric="roc_curve")
```

```python
turicreate.show(evaluate_roc['roc_curve']['fpr'], evaluate_roc['roc_curve']['tpr'])
```

## 13

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
# Loop through word counts to create a classifier for only a few words 
# Created an individual column for each item 
for word in selected_words:
    products[word] = products['kelime_sayaci'].apply(lambda counts: counts.get(word, 0))

products
```

```python
for word in selected_words:
  print("\n ''{}''kelimesi {} kere kullanılmış".format(word, products[word].sum()))
```

Seçili kelimeler arasında en fazla ''great'' kelimesi kullanılmış

## 14

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
features = selected_words
```

```python
selected_words_model = turicreate.logistic_classifier.create(train_data,target='sentiment', features=features,validation_set=test_data)
```

## 15

```python
selected_words_model.coefficients.sort(key_column_names='value', ascending=True)
```

en pozitif ağırlık 'awesome' kelimesinde.

en negatif ağırlık 'horrible' kelimesinde. 

Gayet Anlamlı!
